```python
from base_code import *

valid_data = pd.read_csv('../neurosurgeon35_data/neurosurgeon35.csv', index_col='Respondent ID')
col_name_dict = {'For your selected scenarios to check the explanation from AI, would you mind ranking them regarding your likelihood of checking the explantions?': 'To build and calibrate my trust in this AI',
 'Unnamed: 187': 'To ensure the safety use of the AI',
 'Unnamed: 188': 'To meet the ethical requirements',
 'Unnamed: 189': 'To meet the legal requirements',
 'Unnamed: 190': 'When I doubt about the prediction from AI',
 'Unnamed: 191': 'For a difficult case, when I am not certain',
 'Unnamed: 192': 'To ensure fairness and no biases in the AI model',
 'Unnamed: 193': 'To generate report or patient chart',
 'Unnamed: 194': 'To make Differential Diagnosis',
 'Unnamed: 195': 'Before discussion with my colleagues',
 'Unnamed: 196': 'To learn from AI',
 'Unnamed: 197': 'To make new medical discovery',
 'Unnamed: 198': "To improve my patients' outcomes",
 'Unnamed: 199': 'When I am trading off among multiple objectives for my patient',
 'Unnamed: 200': "To verify AI's decisions"}
```

# 1. Data prep

```python
# get xai_usage columns data
for idx, col in enumerate(valid_data.columns):
    if col =='When are you most likely to check those color map explanations from AI?':
        print(idx, col)
    if col == 'What is your understanding of artificial intelligence (AI)?':
        print(idx, col)
    if col == 'For your selected scenarios to check the explanation from AI, would you mind ranking them regarding your likelihood of checking the explantions?':
        print(idx, col)
```

```python
valid_data = valid_data.rename(columns= col_name_dict)
xai_selection_times = {k:0 for k in col_name_dict.values()}
xai_usage = valid_data.iloc[:, 158:173]
xai_ranking = valid_data.iloc[:, 174:189]
for col in xai_usage.columns:
    usage = xai_usage[col].dropna().unique().tolist()
    selection_times = xai_usage[col].dropna().shape[0]
    if len(usage) >0 and usage[0] in xai_selection_times:
        xai_selection_times[usage[0]] =selection_times
    print(usage, selection_times)
max_rank = 0
scores = dict()
times = dict()
ranking_dict = dict()
xai_ranking
```

```python
for col in xai_ranking.columns:
    ranks = xai_ranking[col].dropna().value_counts().to_dict()
    if len(ranks) >0:
        if max([float(r) for r in ranks.keys()])> max_rank:
            max_rank = max([float(r) for r in ranks.keys()])
    total_rank = sum([v for v in ranks.values()])
    if col in xai_selection_times:
        print(total_rank, xai_selection_times[col])
    print(col, ranks)
    score = sum([(11-float(k))*v for k,v in ranks.items()])
    scores[col] = score
    ranking_dict[col] = ranks
    times[col] = total_rank

print(max_rank)
print(scores)
```

# 2. Report result in table

```python
# reporting in table
for goal, score in dict(sorted(scores.items(), key=lambda item: item[1], reverse=True)).items():
    ranking_sorted = dict(sorted(ranking_dict[goal].items(), key=lambda item: float(item[0]))) 
    ranking_str = ''
    for k,v in ranking_sorted.items():
        ranking_str+= '\\textbf{{{}}} ({}), '.format(k,v)
        
    print('{} & {} & {}\\\\ \\hline'.format(goal, times[goal], ranking_str))

```
